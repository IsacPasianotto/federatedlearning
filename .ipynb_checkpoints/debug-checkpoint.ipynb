```python
data_small_loaded = torch.load('data/brain_cancer_dataset_small.pt')
```

```python
# extract some random indexes 
n_ind = 30 
ind = torch.randint(0, len(data_small_loaded), (n_ind,))
ind
```

```python
# for i in range(n_ind ):
#     img, lab = data[ind[i]]
#     plt.imshow(img[0], cmap='gray')
#     #  1 --> brain glioma
#     #  2 --> meningioma
#     #  3 --> tumor
#     lab_txt = ""
#     if lab == 1.0:
#         lab_txt = 'Brain Glioma'
#     if lab == 2.0:
#         lab_txt = 'Meningioma'
#     if lab == 3.0:
#         lab_txt = 'Tumor'
#     plt.title(lab_txt)
#     plt.show()

for i in range(n_ind ):
    img, lab = data_small_loaded[ind[i]]
    plt.imshow(img[0], cmap='gray')
    #  1 --> brain glioma
    #  2 --> meningioma
    #  3 --> tumor
    lab_txt = ""
    if lab == 1.0:
        lab_txt = 'Brain Glioma'
    if lab == 2.0:
        lab_txt = 'Meningioma'
    if lab == 3.0:
        lab_txt = 'Tumor'
    plt.title(lab_txt)
    plt.show()
```
